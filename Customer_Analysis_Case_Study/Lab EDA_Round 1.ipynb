```python
def clean_gender(x):
    if x in ['F','female','Femal']:
        return 'Female'
    elif x in ['M','Male']:
        return 'Male'
    else:
        return 'U'
```

```python
ca_df['Gender'] = list(map(clean_gender,ca_df['Gender'])) 
```
